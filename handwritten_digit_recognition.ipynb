```python
def get_featurs_with_zero_value(data_frame):
    """ Function to get the list of features which have only 0's
    """
    drop_feature_list = []
    for df in data_frame:
        if data_frame[df].nunique()  == 1 and data_frame[df].unique()[0] == 0:
            drop_feature_list.append(df)
    return drop_feature_list
```

** 3a. k-Nearest Neighbour**
1. Non-parametric model i.e. no mathematicl equation which describes the relationship between X and y
2. Create a database of "X" and corresponding values of "y" using training data
3. When new "X" is given for classification 
    - euclidean (minkowski) distance with respect to all training data point is calculated
    - first 'k' closest instance of training data is taken
    - from this 'k' closest instance take the majority 'y' as the label for 

- It gives good accuracy but does not scale well as distance with respect to all images needs to be calcualted

http://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html

```python
## Using sci-kit learn Nearest Neighbour Classifer

from sklearn.neighbors import KNeighborsClassifier

st = time()
knc = KNeighborsClassifier()
knc.fit(X_train, y_train)
time_knc = time() - st
```

**3b. Logistic Regression**

1. Parametric model i.e. we assume some mathematical relationship between "X" and "y"
    - Between pixel intensity and classes (digits)
2. We assume Linear Decision Boundary (Mathemtical) relationship between "X" and "y" is assumed
   $$ \hat{y} = w_0 + w_1 * x_1 + w_2*x_2 + .....+ w_n * x_n $$
3. $\hat{y}$ value then can be used to decide between two classes
    - this will not give very good decision boundary 
    
    <img src="images/classification_problem_v1.png">
    
    - so, we convert value of $\hat{y}$ to probability using link function, mostly sigmoid function 
        $$S(\hat{y}) = \frac{1}{1 + exp(-\hat{y})}$$
        
    <img src="images/Logistic-curve.svg.png">
    
    - We put a cut-off on porbability value and derive the class information 
        $$ if\hspace{2mm}S(\hat{y}) > threshold(0.5)\hspace{2mm} then\hspace{2mm} class\hspace{1mm} 1\hspace{2mm} else\hspace{2mm} class\hspace{1mm} 2$$
4. This works perfectly well in case of binary classification (class 1 vs. class 2)
5. For multiclass we perform one vs. rest i.e. 
    - probability of class 1 vs. not being class 1
    - probability of class 2 vs. not being class 2, etc.
6. So, we create as many model as we have classes
    - in case of hand written digit we will learn 10 models as we have 10 digit
    - so, Number of parameters we will be equal to No. of Fearures * 10 
7. We get 10 probability values for each "X". Digit is one which gives maximum probability 

    $$ digit\hspace{2mm}= argmax\hspace{2mm} (prob)$$
    
http://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html

```python
# Using sci-kit learn Logistic Regression - Multiclass Classification

from sklearn.linear_model import LogisticRegression

st = time()
lr = LogisticRegression()
lr.fit(X_train, y_train)
time_lr = time() - st
```

**3c. Decision Trees**
1. Non Parametric Model
2. Create binary decision trees by splitting on "X" or pixel intensity
3. Keep splitting till we reach some kind of stopping criterion like - 
    - leaf is pure i.e. only one class is present
    - no more data to split or amount of data present in leaf is less than the threshold level
    - maximum depth is reached, etc.. 
4. Data present in the leaf node gives the probability values for a class 

<img src="images/decision_tree.png">

http://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html

```python
def predict_for_any_input_image(image_file):
    
    from skimage import io, transform
    
    img = io.imread(image_file)
    print("\n Size of original image ", img.shape)
    print("\n Expected size is 28 * 28, so re-sizing")
    plt.subplot(121)
    plt.imshow(img)
    plt.title("Original Image")
    
    
    img = transform.resize(img, (28,28)).mean(axis=2)
    img = 1- img
    
    plt.subplot(122)
    plt.imshow(img, cmap=plt.cm.gray)
    plt.title("Transformed image")
    plt.show()
    print("Prediciton:: K-Nearest Neighbour digit is:: ", 
          int(knc.predict(img.ravel().reshape(1,-1))[0]))
    
    print("Prediciton:: Logistic Classifer digit is:: ", 
          int(lr.predict(img.ravel().reshape(1,-1))[0]))
    
    print("Prediciton:: Decision Tree Classifer digit is:: ", 
          int(dtc.predict(img.ravel().reshape(1,-1))[0]))
    
    print("Prediciton:: Random Forest digit is:: ", 
          int(rf.predict(img.ravel().reshape(1,-1))[0]))
```
